```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
from json import dumps
import json
```

```python
producer = KafkaProducer(bootstrap_servers=[':9092'], #change ip here
                         value_serializer=lambda x: 
                         dumps(x).encode('utf-8'))
```

```python
producer.send('Pipelinetest', value={'name':'HarshVictorChalla'})
```

```python
df = pd.read_csv("Historical_Stock_Data.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Date'].min(), df['Date'].max()
```

```python
# Send each row to the Kafka topic, simulating real-time data streaming
for index, row in df.iterrows():
    dict_stock = row.to_dict()
    producer.send('Pipelinetest', value=dict_stock)
    sleep(1)

    
```

```python
producer.flush() #clear data from kafka server
```
